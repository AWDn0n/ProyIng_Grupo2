```python
import pandas as pd
import plotly.express as px
import plotly.graph_objects as go
```

```python
train = pd.read_csv('data/train_cleaned.csv')
test = pd.read_csv('data/testing_set.csv')
train.drop(columns=['valid','outlier','valid2'], inplace=True)
```

```python
# plot candlestick of train data
import plotly.graph_objects as go

label_index = 0 # replace 0 with the index of the label you want to display

fig = go.Figure(data=[go.Candlestick(x=train['Time'],
                open=train['Open'],
                high=train['High'],
                low=train['Low'],
                close=train['Close'],
                # add color parameter
                increasing=dict(line=dict(color='green')),
                decreasing=dict(line=dict(color='red')))])

# Add text annotation for label value
fig.add_annotation(x=train['Time'][label_index], y=train['High'][label_index],
                   text=f"Label value: {train['label'][label_index]}",
                   showarrow=True, arrowhead=1)

fig.show()
```

```python
# lineplot of train data with label
fig = px.line(train, x='Time', y='Close', color='label')
fig.show()
```

## Method 1: Randomly assign values

```python
# create a column called label for test data
# 0 if the price goes down, 1 if the price goes up
test['label'] = test['Close'].diff().apply(lambda x: 1 if x > 0 else 0)
```

```python
# Create a column called label for test data
# set 0 or 1 randomly with equal probability
import random

test['label'] = [random.randint(0, 1) for _ in range(len(test))]
```

```python
# make a json with the following format
# {{"target":{"0":0,...,"n":0}}}
# where n is the test_idx in the test data
# and 0 is the label value for each row
json = test[['label','test_idx']].to_json()
# rename label to target
json = json.replace('label','target')
```

```python
# save json to a file
with open('data/test_labels_rng.json', 'w') as f:
    f.write(json)
```
