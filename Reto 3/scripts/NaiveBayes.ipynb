```python
import pandas as pd

# Read the training data, train_idx is the index of the training data
train = pd.read_csv('../data/train_cleaned.csv', parse_dates=True)
train.head()
```

```python
# Drop train_idx, Time, valid, outlier and valid2 columns
train.drop(['train_idx','Time','valid', 'outlier', 'valid2'], axis=1, inplace=True)
train.head()
```

```python
# Import scikit-learn libraries for a naive-bayes classifier
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score

# Create a naive-bayes classifier
gnb = GaussianNB()

# Train the classifier with the training data
gnb.fit(train.drop('label', axis=1), train['label'])
```

```python
# Read the test data, test_idx is the index of the test data
test = pd.read_csv('../data/testing_set.csv')
test.head()
```

```python
# Drop test_idx and Time columns
test.drop(['test_idx','Time'], axis=1, inplace=True)
test.head()
```

```python
# Predict the labels of the test data
pred = gnb.predict(test)
```

```python
# Create a dataframe with the test_idx and the predicted labels
pred_df = pd.DataFrame({'test_idx':test.index, 'label':pred})
pred_df.head()
```

```python
# Save the dataframe as a JSON file, the test_idx column values is the index of the test data
pred_df.to_json('../data/predsNB.json', index=True)
```
