```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
train_data = pd.read_csv('./ProyIng_Grupo2/Reto3//data/train_cleaned.csv')
test_data = pd.read_csv('./ProyIng_Grupo2/Reto3/data/testing_set.csv')
```

```python
train_data=train_data.drop(['Time'], axis=1)
test_data=test_data.drop(['Time'], axis=1)
```

```python
import pandas as pd
from sklearn.ensemble import AdaBoostClassifier
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
from sklearn.model_selection import cross_val_score







```

```python
# Separar las características (X_train) y las etiquetas (y_train)
X_train = train_data.drop('label', axis=1)  # Ajusta 'target_column' al nombre de tu columna objetivo
y_train = train_data['label']  # Ajusta 'target_column' al nombre de tu columna objetivo

# Crear un clasificador AdaBoost
ada_boost = AdaBoostClassifier(n_estimators=50, random_state=42)

# Realizar validación cruzada y obtener la precisión media
cv_scores = cross_val_score(ada_boost, X_train, y_train, cv=5)  # Ajusta el número de "folds" (pliegues) según tus necesidades
mean_accuracy = cv_scores.mean()

print("Precisión promedio: {:.2f}%".format(mean_accuracy * 100))
```
