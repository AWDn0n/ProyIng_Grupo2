```python
import tensorflow as tf
from keras import layers, models
from keras.preprocessing.image import ImageDataGenerator

import pandas as pd
import numpy as np
from keras.models import Sequential
from keras.layers import Conv2D, MaxPooling2D, Flatten, Dense
```

```python
# initialize the number of epochs to train for, initial learning rate,
# and batch size
EPOCHS = 25
INIT_LR = 1e-3
BS = 32
```

```python
# Load the dataset
data_train = pd.read_csv('data/train.csv')

# Define image dimensions
img_width, img_height = 259, 194

# Define number of classes
num_classes = 3

# Define input shape
input_shape = (img_height, img_width, 3)

# Define the CNN model
model = Sequential()
model.add(Conv2D(32, (3, 3), activation='relu', input_shape=input_shape))
model.add(MaxPooling2D((2, 2)))
model.add(Conv2D(64, (3, 3), activation='relu'))
model.add(MaxPooling2D((2, 2)))
model.add(Conv2D(128, (3, 3), activation='relu'))
model.add(MaxPooling2D((2, 2)))
model.add(Flatten())
model.add(Dense(128, activation='relu'))
model.add(Dense(num_classes, activation='softmax'))

# Compile the model
model.compile(optimizer='adam',
              loss='categorical_crossentropy',
              metrics=['accuracy'])
```

```python
# Define training and validation data
train_images = []
train_labels = []
for i, row in data_train.iterrows():
    print(f'starting image {i} of {len(data_train)}')
    img_path = row['path_img']
    label = row['label']
    img = tf.keras.preprocessing.image.load_img(img_path, target_size=(img_height, img_width))
    img_array = tf.keras.preprocessing.image.img_to_array(img)
    train_images.append(img_array)
    train_labels.append(label)
train_images = np.array(train_images)
train_labels = tf.keras.utils.to_categorical(train_labels, num_classes=num_classes)
```

```python
# Fit the model
history = model.fit(train_images, train_labels, epochs=10, batch_size=32)
```

```python
# Load the test set data
test_df = pd.read_csv('data/test_split.csv')

# Preprocess the test images
X_test = []
for i, file_path in enumerate(test_df['path_img']):
    # Load the image
    img = cv2.imread(file_path)
    # Resize the image to match the dimensions of the training images
    img_resized = cv2.resize(img, (img_width, img_height))
    # Normalize the pixel values
    img_normalized = img_resized / 255.0
    # Append the processed image to the list of test images
    X_test.append(img_normalized)

# Convert the test set to numpy arrays
X_test = np.array(X_test)
y_test = np.array(test_df['label'])

# Evaluate the model on the test set
test_loss, test_acc = model.evaluate(X_test, y_test)
print('Test loss:', test_loss)
print('Test accuracy:', test_acc)
```

```python
# Define image dimensions
img_width, img_height = 194, 259

# Define number of classes
num_classes = 8

# Define batch size
batch_size = 32

# Define number of epochs
epochs = 10

# Define paths to training and validation data
train_data_dir = 'data/test_split.csv'
validation_data_dir = 'data/validation.csv'

# Data preprocessing and augmentation
train_datagen = ImageDataGenerator(rescale=1./255,
                                   rotation_range=20,
                                   width_shift_range=0.1,
                                   height_shift_range=0.1,
                                   zoom_range=0.1,
                                   horizontal_flip=True)

test_datagen = ImageDataGenerator(rescale=1./255)

train_generator = train_datagen.flow_from_directory(train_data_dir,
                                                    target_size=(img_width, img_height),
                                                    batch_size=batch_size,
                                                    class_mode='categorical')

validation_generator = test_datagen.flow_from_directory(validation_data_dir,
                                                        target_size=(img_width, img_height),
                                                        batch_size=batch_size,
                                                        class_mode='categorical')

# Define the model
model = models.Sequential()

# Add convolutional layers
model.add(layers.Conv2D(32, (3, 3), activation='relu', input_shape=(img_width, img_height, 3)))
model.add(layers.MaxPooling2D((2, 2)))
model.add(layers.Conv2D(64, (3, 3), activation='relu'))
model.add(layers.MaxPooling2D((2, 2)))
model.add(layers.Conv2D(128, (3, 3), activation='relu'))
model.add(layers.MaxPooling2D((2, 2)))

# Add fully connected layers
model.add(layers.Flatten())
model.add(layers.Dense(128, activation='relu'))
model.add(layers.Dense(num_classes, activation='softmax'))

# Compile the model
model.compile(loss='categorical_crossentropy',
              optimizer='adam',
              metrics=['accuracy'])
```

```python
# Evaluate the model
test_loss, test_acc = model.evaluate(validation_generator)

# Print the test accuracy
print('Test accuracy:', test_acc)
```

```python
# Train the model
history = model.fit(train_generator,
                    epochs=epochs,
                    validation_data=validation_generator)
```
